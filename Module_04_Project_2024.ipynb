## Gradient Descent

```python
########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)## [REQUIRED]
    new_x = previous_x - step_multiplier * gradient## [REQUIRED]

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x)) ## [REQUIRED])

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
## TASK-5 : Run gradient descent 3 times
import numpy as np

def gradient_descent(x, y, learning_rate, epochs):
    m = len(x)
    theta = np.zeros((2, 1))  # Initializing parameters with zeros

    for epoch in range(epochs):
        # Hypothesis function
        h = np.dot(x, theta)

        # Calculate the cost function
        cost = (1 / (2 * m)) * np.sum((h - y)**2)

        # Calculate the gradient
        gradient = (1 / m) * np.dot(x.T, (h - y))

        # Update parameters using the gradient and learning rate
        theta -= learning_rate * gradient

        if epoch % 100 == 0:
            print(f'Epoch {epoch}, Cost: {cost}')

    return theta

# Generate some sample data
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# Add bias term to x
x_b = np.c_[np.ones((100, 1)), x]

# Set hyperparameters
learning_rate = 0.01
epochs = 1000

# Run gradient descent
theta = gradient_descent(x_b, y, learning_rate, epochs)

print(f'Optimal Parameters: {theta}')
```

```python
## Task -6 :plot two morelearing rates:mid_gamma(0.001) and high_gamma(0.002) give code



import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, learning_rate, epochs):
    m = len(x)
    theta = np.zeros((2, 1))
    cost_history = []

    for epoch in range(epochs):
        h = np.dot(x, theta)
        cost = (1 / (2 * m)) * np.sum((h - y) ** 2)
        gradient = (1 / m) * np.dot(x.T, (h - y))
        theta -= learning_rate * gradient
        cost_history.append(cost)

    return theta, cost_history

def plot_cost_over_epochs(learning_rate, epochs, x, y):
    x_b = np.c_[np.ones((len(x), 1)), x]

    theta_low_gamma, cost_history_low_gamma = gradient_descent(x_b, y, learning_rate=0.001, epochs=epochs)
    theta_mid_gamma, cost_history_mid_gamma = gradient_descent(x_b, y, learning_rate=0.01, epochs=epochs)
    theta_high_gamma, cost_history_high_gamma = gradient_descent(x_b, y, learning_rate=0.02, epochs=epochs)

    plt.plot(range(epochs), cost_history_low_gamma, label=f'Low Gamma ({learning_rate})')
    plt.plot(range(epochs), cost_history_mid_gamma, label=f'Mid Gamma (0.01)')
    plt.plot(range(epochs), cost_history_high_gamma, label=f'High Gamma (0.02)')

    plt.xlabel('Epochs')
    plt.ylabel('Cost')
    plt.title(f'Gradient Descent with Different Learning Rates')
    plt.legend()
    plt.show()

# Generate some sample data
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# Set hyperparameters
epochs = 100

# Plot cost over epochs for different learning rates
plot_cost_over_epochs(learning_rate=0.001, epochs=epochs, x=x, y=y)
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```
